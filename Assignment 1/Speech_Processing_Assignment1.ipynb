```python
!pip install librosa==0.10.0
```

```python
#Load the audio
import librosa
audio_path = "words.mp3"
x ,sr = librosa.load(audio_path)

#x : array
#sr : sampling rate in Hz
```

```python
import IPython.display as ipd
ipd.Audio(audio_path)
```

```python
import matplotlib.pyplot as plt
import librosa.display
```

```python
#visualizing the audio
plt.figure(figsize=(14,5))
librosa.display.waveshow(x, sr = sr)
```

```python
#Spectogram
X = librosa.stft(x) #audio to short term fourier transform
Xdb = librosa.amplitude_to_db(abs(X)) #amplitude at a given time
```

```python
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb, sr = sr, x_axis = 'time', y_axis='hz')
plt.colorbar()
```

```python
import speech_recognition as sr
```

```python
r = sr.Recognizer()
```

```python
#using record to capture data from file
sample_audio = sr. AudioFile('English.wav')
print(type(sample_audio))
with sample_audio as source:
  audio = r.record(source)
```

```python
r.recognize_google(audio)
```

```python
sample_audio2 = sr. AudioFile('Hindi.wav')
print(type(sample_audio2))
with sample_audio2 as source2:
  audio2 = r.record(source2)
```

```python
r.recognize_google(audio2, language='hi')
```

### Output for Question 4: Both languages are recognized correctly and they are displayed properly
